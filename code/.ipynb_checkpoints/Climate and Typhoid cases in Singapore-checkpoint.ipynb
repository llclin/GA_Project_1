## Background

According to the [Meteorological Services Singapore](http://www.weather.gov.sg/climate-climate-of-singapore/#:~:text=Singapore%20is%20situated%20near%20the,month%2Dto%2Dmonth%20variation.), Singapore has typical tropical climate with adundant rainfall, high and uniform temperatures and high humidity all year round, since its situated near the equator. There are many factors that help us understand the climate of a country and in this project we are going to look into a few, especially rainfall.

Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods.  The **Northeast Monsoon** occurs from December to early March, and the **Southwest Monsoon** from June to September.

The major weather systems affecting Singapore that can lead to heavy rainfall are:

-Monsoon surges, or strong wind episodes in the Northeast Monsoon flow bringing about major rainfall events;

-Sumatra squalls, an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of us;

-Afternoon and evening thunderstorms caused by strong surface heating and by the sea breeze circulation that develops in the afternoon.

Singapore’s climate station has been located at several different sites in the past 140 years. The station had been decommissioned at various points in the past due to changes to local land use in the site’s vicinity, and had to be relocated. Since 1984, the climate station has been located at **Changi**.

There are other metrics of climate such as temperature, humidity, sun shine duration, wind speed, cloud cover etc. All the dataset used in the project comes from [data.gov.sg](data.gov.sg), as recorded at the Changi climate station

### Coding Challenges

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:

def cal_mean(mylist):
    total = 0
    for num in mylist:
        total += num 
    my_avg = total/len(mylist)
        
    return my_avg
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:

def cal_std(mylist):
    my_avg = cal_mean(mylist)
    sub_list = []
    for i in range(len(mylist)):
        sub_list.append((mylist[i]-my_avg)**2)
    summation_total = 0
    for num in sub_list:
        summation_total += num
    my_std = ((1/len(mylist))*summation_total)**0.5
    return my_std
```

```python
#1. Display the data: print the first 5 rows of each dataframe to your Jupyter notebook.

df_monthly_total = pd.read_csv("../data/RainfallMonthlyTotal.csv")
print(df_monthly_total.head())

df_monthly_num_rain_days = pd.read_csv("../data/RainfallMonthlyNumberofRainDays.csv")
print(df_monthly_num_rain_days.head())

df_monthly_mean_temp = pd.read_csv("../data/SurfaceAirTemperatureMonthlyMean.csv")
print(df_monthly_mean_temp .head())

df_monthly_highest_rainfall = pd.read_csv("../data/RainfallMonthlyHighestDailyTotal.csv")
print(df_monthly_highest_rainfall.head())

df_monthly_sunshine = pd.read_csv("../data/SunshineDurationMonthlyMeanDailyDuration.csv")
print(df_monthly_sunshine.head())

df_monthly_humidity = pd.read_csv("../data/RelativeHumidityMonthlyMean.csv")
print(df_monthly_humidity.head())

df_infectious_disease = pd.read_csv("../data/WeeklyInfectiousDiseaseBulletinCases.csv")
print(df_infectious_disease.head())
```

```python
#2. Check for missing values and datatype.
#3. Check for any obvious issues with the observations.
print(df_monthly_total.info())
print(df_monthly_num_rain_days.info())
print(df_monthly_mean_temp.info()) #df_montly_mean has one less observation compared to the other rain related data.
print(df_monthly_highest_rainfall.info())
print(df_monthly_sunshine.info())
print(df_monthly_humidity.info())
print(df_infectious_disease.info()) #we need to filter for "typhoid" and group the cases by month before merging.
```

```python
#3. Drop unnecessary rows (if needed).
#4. Merge dataframes that can be merged.
#5. Since different climate metrics are in month format, you can merge them into one single dataframe for easier analysis
#6. Perform any additional cleaning that you feel is necessary.


#Given that all the climate related datasets of Singapore have a common field "month", we can combine them all together.
df_climate_data = pd.merge(df_monthly_total,df_monthly_num_rain_days, on = "month", how = "left")
df_climate_data = pd.merge(df_climate_data,df_monthly_mean_temp, on = "month", how = "left")
df_climate_data = pd.merge(df_climate_data, df_monthly_highest_rainfall, on = "month", how = "left")
df_climate_data = pd.merge(df_climate_data, df_monthly_sunshine, on = "month", how = "left")
df_climate_data = pd.merge(df_climate_data,df_monthly_humidity, on = "month", how = "left")
#We will also need to remove the rows with any empty observation for the climate related dataset.
df_climate_data = df_climate_data.dropna()
#As we may need to analyse the data by years or by the months, we will need to transform the data as well to get the "month_of_year" and "year" column.
df_climate_data["year-month"] = df_climate_data['month'].str.split("-")
df_climate_data["year"] = df_climate_data["year-month"].str.get(0)
df_climate_data["month_of_year"] = df_climate_data["year-month"].str.get(1)
df_climate_data = df_climate_data.drop(columns = ["year-month"])

display(df_climate_data)
#For this study as we are not using the year and month for any datetime calculation, it is not necessary to change it to a datetime format. 
display(df_climate_data.info())

#For our subsequent analysis of the climate data, we would prefer to have full year of data for same parity comparison.
#As we do not have the full year data for 2023, we will be removing it.
df_climate_data = df_climate_data[df_climate_data["year"]!='2023']

#We save the merged climate data into a csv file for future use on analysing the climate changes in Singapore.
df_climate_data.to_csv("climate_data.csv")
```

```python

#The following codes is to transform the infectious disease dataset to filter for typhoid disease and to group cases by month.
df_infectious_disease['end_of_week'] = df_infectious_disease['epi_week'].apply(lambda x: Week.fromstring(x).enddate())
df_infectious_disease['month'] = df_infectious_disease['end_of_week'].astype(str).str[:7]
df_infectious_disease_pivot = pd.pivot_table(df_infectious_disease, values='no._of_cases', index=['month'], columns=['disease'], aggfunc=np.sum)
df_disease = df_infectious_disease_pivot.reset_index()
df_typhoid = df_disease.loc[:,["month","Typhoid"]]
df_typhoid = df_typhoid.rename(columns={"month":"month", "Typhoid": "typhoid"})
display(df_typhoid)
```

```python
#We merge by taking all the available data on typhoid cases and map it with climate data for our analysis.
df = pd.merge(df_typhoid, df_climate_data, how = "left", on = "month")


#We then create a new dataframe to categorise the month_of_year based on the type of monsoon seasons that Singapore has.
#Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods. 
#The Northeast Monsoon occurs from December to early March, and the Southwest Monsoon from June to September.
#Northeast Monsoon will be represented by"NE", Southwest Monsoon will be represented by "SW". 
#For the remaining months not under the monsoon season will be representeed by "NA".

monsoon_df = pd.DataFrame(np.array([['01',"NE"],['02',"NE"],['03',"NE"],['12',"NE"],['06',"SW"],['07','SW'],['08',"SW"],['09',"SW"],['04',"NA"],['05',"NA"],['10',"NA"],['11',"NA"]]),
                          columns = ["month_of_year","monsoon_type"])


df_full = pd.merge(df,monsoon_df,on = "month_of_year", how = "left")

display(df_full.head())
print(df_full.info())

#As we are interested in the past 5 years' data on analysing the relationship in recent climate and 
#the number of locally reported typhoid cases,we will be filtering the period to look at the cases from 2018 to 2022. 
required_yr = ["2018","2019","2020","2021","2022"]
df_final = df_full[df_full["year"].isin(required_yr)]

display(df_final.head())
print(df_final.info())

#We save the dataset of our interest into a csv file for future use to analyse the relationship between Singapore's
#climate and the number of Typhoid cases.
df_final.to_csv("typhoid_and_climate_data.csv")
```

```python
#Plotting heatmap of all numeric columns to see if there is strong correlation. 
plt.figure(figsize=(15,10))
sns.heatmap(df_final.corr(), annot=True).set(title = "Correlation between all numeric columns of the climate and typhoid cases data")
```

Based on the heatmap showing the correlation between all the numeric columns of the climate and typhoid cases data (i.e. df_final), we are able to detect the following correlations:

1) There is moderate positive correlation between the total rainfall and the number of rainy days (i.e. correlation of 0.57). This confirms our earlier suspicion that the total rainfall and number of rainy days do not hold strong positive correlation. 


2) There also exists slight negative correlation between the average relative humidity and the number of typhoid cases and (i.e. correlation of -0.35). This meant that the higher the average relative humidity, we should see lower number of typhoid cases.  

However, based on the current data, we are not able to conclude that there is moderate or strong correlations between typhoid cases with the various climate metrics recorded. We could then consider segmenting the data by the type of monsoon which would have differing climate metrics to see whether there are specific monsoon seasons where the number of typhoid cases has a stronger correlation with the specific climate metrics recorded.

```python
# Code:

#We now split the data by the type of monsoon periods to study the data further.
df_SW = df_final[df_final["monsoon_type"]=="SW"]
df_NE = df_final[df_final["monsoon_type"]=="NE"]
df_NA = df_final[df_final["monsoon_type"]=="NA"]
```

```python
#With the splitted data, we plot the various heatmaps to look out for the correlation for the different monsoon types.
plt.figure(figsize=(15,10))
sns.heatmap(df_SW.corr(), annot=True).set(title="Correlation between all numeric columns of the climate and typhoid cases data during Southwest monsoon period")
plt.figure(figsize=(15,10))
sns.heatmap(df_NE.corr(), annot=True).set(title="Correlation between all numeric columns of the climate and typhoid cases data during Northeast monsoon period")
plt.figure(figsize=(15,10))
sns.heatmap(df_NA.corr(), annot=True).set(title="Correlation between all numeric columns of the climate and typhoid cases data during non monsoon period")
```

After we segment the data by the Southwest monsoon period, Northeast monsoon period and non-monsoon period, we can see from the heatmaps that during the Southwest monsoon period, there is higher moderate negative correlation of -0.59 between the number of typhoid cases and the average relative humidity, compared to when we plot it against the full data. There is also a higher moderate negative correlation of -0.47 between the number of typhoid cases and the number of rainy days, compared to when we plot it against the full data. 

On the contrary, for the Northeast monsoon period and non monsoon period, there is much weaker negative correlation between the number of typhoid cases and the average relative humidity, compared to when we plot it against the full data.

With the above, we can see that for the months of Jun to Sep of the calendar year which happens to be the Southwest monsoon period, when the average relative humidity decreases during this period, we generally see more typhoid cases reported in Singapore.

Based on the boxplot, we can see that the median of the average relative humidity during the Southwest monsoon period is lower than the other two groups.
